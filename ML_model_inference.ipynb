# Fibroscan

```python
import pandas as pd
from pycaret.regression import *

df = pd.read_csv("data/fibroscan_predict_df.csv")
df.loc[:, "FIB4"] = df.apply(lambda x: (x["age"] * x["AST"]) / (x['PLT'] * np.sqrt(x['ALT']) + 1e-8), axis=1)
df.loc[:, "APRI"] = df.apply(lambda x: (x["AST"] / 34) * 100 / (x['PLT'] + 1e-8), axis=1)
df = df.drop("ID", axis=1)
df.kPa_fib = df.kPa_fib.map(lambda x: 16 if x > 16 else x)
df.kPa_fib = df.kPa_fib.map(lambda x: 4 if x < 4 else x)
reg = setup(data=df, target='kPa_fib', fold=5, session_id=42)
```

```python
lr = create_model('lr')
tuned_xgb = tune_model(
    lr, optimize='RMSE', n_iter=100, 
    search_library='optuna', choose_better=True)
```

```python
dt = create_model('dt')
tuned_dt = tune_model(
    dt, optimize='RMSE', n_iter=100, 
    search_library='optuna', choose_better=True)
```

```python
lightgbm = create_model('lightgbm')
tuned_lightgbm = tune_model(
    lightgbm, optimize='RMSE', n_iter=100, 
    search_library='optuna', choose_better=True)
```

```python
best_model = finalize_model(lightgbm)
save_model(best_model, 'weights/fibroscan_lightgbm')
```

# MRE

```python
import pandas as pd
from pycaret.regression import *

df = pd.read_csv("data/mre_predict_df.csv")
df.kPa_mre = df.kPa_mre.map(lambda x: 7 if x > 7 else x)
df.kPa_mre = df.kPa_mre.map(lambda x: 2 if x < 2 else x)
df.loc[:, "FIB4"] = df.apply(lambda x: (x["age"] * x["AST"]) / (x['PLT'] * np.sqrt(x['ALT']) + 1e-8), axis=1)
df.loc[:, "APRI"] = df.apply(lambda x: (x["AST"] / 34) * 100 / (x['PLT'] + 1e-8), axis=1)
df = df.drop("ID", axis=1)
reg = setup(data=df, target='kPa_mre', fold=5, session_id=42)
```

```python
lr = create_model('lr')
tuned_xgb = tune_model(
    lr, optimize='RMSE', n_iter=100, 
    search_library='optuna', choose_better=True)
```

```python
dt = create_model('dt')
tuned_dt = tune_model(
    dt, optimize='RMSE', n_iter=100, 
    search_library='optuna', choose_better=True)
```

```python
lightgbm = create_model('lightgbm')
tuned_lightgbm = tune_model(
    lightgbm, optimize='RMSE', n_iter=100, 
    search_library='optuna', choose_better=True)
```

```python
best_model = finalize_model(lightgbm)
save_model(best_model, 'weights/mre_lightgbm')
```
