```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import scipy.stats
import numpy as np


def draw_plot_fibroscan(mean, std=3.48, x_min=0, x_max=30):

    x = np.linspace(x_min, x_max, 100)
    y = scipy.stats.norm.pdf(x, mean, std)

    plt.plot(x, y, color='black')

    #----------------------------------------------------------------------------------------#
    # fill area 1

    pt1 = mean + std
    plt.plot([pt1 ,pt1 ],[0.0,scipy.stats.norm.pdf(pt1 ,mean, std)], color='black')

    pt2 = mean - std
    plt.plot([pt2 ,pt2 ],[0.0,scipy.stats.norm.pdf(pt2 ,mean, std)], color='black')

    ptx = np.linspace(pt1, pt2, 10)
    pty = scipy.stats.norm.pdf(ptx,mean,std)

    plt.fill_between(ptx, pty, color='#0b559f', alpha=1.0)

    #----------------------------------------------------------------------------------------#
    # fill area 2

    pt1 = mean + std
    plt.plot([pt1 ,pt1 ],[0.0,scipy.stats.norm.pdf(pt1 ,mean, std)], color='black')

    pt2 = mean + 2.0 * std
    plt.plot([pt2 ,pt2 ],[0.0,scipy.stats.norm.pdf(pt2 ,mean, std)], color='black')

    ptx = np.linspace(pt1, pt2, 10)
    pty = scipy.stats.norm.pdf(ptx,mean,std)

    plt.fill_between(ptx, pty, color='#2b7bba', alpha=0.7)

    #----------------------------------------------------------------------------------------#
    # fill area 3

    pt1 = mean - std
    plt.plot([pt1 ,pt1 ],[0.0,scipy.stats.norm.pdf(pt1 ,mean, std)], color='black')

    pt2 = mean - 2.0 * std
    plt.plot([pt2 ,pt2 ],[0.0,scipy.stats.norm.pdf(pt2 ,mean, std)], color='black')

    ptx = np.linspace(pt1, pt2, 10)
    pty = scipy.stats.norm.pdf(ptx,mean,std)

    plt.fill_between(ptx, pty, color='#2b7bba', alpha=0.7)

    #----------------------------------------------------------------------------------------#
    # fill area 4

    pt1 = mean + 2.0 * std
    plt.plot([pt1 ,pt1 ],[0.0,scipy.stats.norm.pdf(pt1 ,mean, std)], color='black')

    pt2 = mean + 3.0 * std
    plt.plot([pt2 ,pt2 ],[0.0,scipy.stats.norm.pdf(pt2 ,mean, std)], color='black')

    ptx = np.linspace(pt1, pt2, 10)
    pty = scipy.stats.norm.pdf(ptx,mean,std)

    plt.fill_between(ptx, pty, color='#539ecd', alpha=0.4)

    #----------------------------------------------------------------------------------------#
    # fill area 5

    pt1 = mean - 2.0 * std
    plt.plot([pt1 ,pt1 ],[0.0,scipy.stats.norm.pdf(pt1 ,mean, std)], color='black')

    pt2 = mean - 3.0 * std
    plt.plot([pt2 ,pt2 ],[0.0,scipy.stats.norm.pdf(pt2 ,mean, std)], color='black')

    ptx = np.linspace(pt1, pt2, 10)
    pty = scipy.stats.norm.pdf(ptx,mean,std)

    plt.fill_between(ptx, pty, color='#539ecd', alpha=0.4)

    #----------------------------------------------------------------------------------------#
    # fill area 6

    pt1 = mean + 3.0 * std
    plt.plot([pt1 ,pt1 ],[0.0,scipy.stats.norm.pdf(pt1 ,mean, std)], color='black')

    pt2 = mean + 10.0 *std
    plt.plot([pt2 ,pt2 ],[0.0,scipy.stats.norm.pdf(pt2 ,mean, std)], color='black')

    ptx = np.linspace(pt1, pt2, 10)
    pty = scipy.stats.norm.pdf(ptx,mean,std)

    plt.fill_between(ptx, pty, color='#89bedc', alpha=0.1)

    #----------------------------------------------------------------------------------------#
    # fill area 7

    pt1 = mean - 3.0 * std
    plt.plot([pt1 ,pt1 ],[0.0,scipy.stats.norm.pdf(pt1 ,mean, std)], color='black')

    pt2 = mean - 10.0 * std
    plt.plot([pt2 ,pt2 ],[0.0,scipy.stats.norm.pdf(pt2 ,mean, std)], color='black')

    ptx = np.linspace(pt1, pt2, 10)
    pty = scipy.stats.norm.pdf(ptx,mean,std)

    plt.fill_between(ptx, pty, color='#89bedc', alpha=0.1)
    
    #----------------------------------------------------------------------------------------#

#     plt.grid()
    
    plt.vlines(mean, 0, 0.115, color="k")
    
    plt.xlim(x_min, x_max)
    plt.ylim(0, 0.20)
    plt.yticks(color='w')

    plt.xlabel('predicted liver stiffness (kPa: fibroscan)')
    
    left_1_sd = mean - std
    if left_1_sd < 0:
        left_1_sd = 0.
    
    left_2_sd = mean - 2 * std
    if left_2_sd < 0:
        left_2_sd = 0.
    
    right_1_sd = mean + std 
    if right_1_sd > x_max:
        right_1_sd = x_max
        
    right_2_sd = mean + 2 * std 
    if right_2_sd > x_max:
        right_2_sd = x_max
    
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    plt.text(19, 0.155,
        f"Mean: {np.round(mean, 2)}\n68% CI: {left_1_sd}-{right_1_sd}\n95% CI: {left_2_sd}-{right_2_sd}", 
        bbox=props
    )
    
    
    plt.show()
    
draw_plot_fibroscan(12.0)
```
